```python
from ultralytics import YOLO
import cv2
```

```python
model = YOLO('yolov8n.pt')
results = model(source="D:\\ComputerVision\\Practical Exam Task\\Detection\\video.mp4", stream=True, show=True, conf=0.4, save=True)

for result in results:
    frame = result.plot()  
    cv2.imshow("Processed Frame", frame) 
```
